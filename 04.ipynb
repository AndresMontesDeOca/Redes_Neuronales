```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder

# Cargar los datos desde el archivo CSV
# data = pd.read_csv('/mnt/data/Semillas.csv')

# Inspeccionar los datos
display(data.head())

# Asumir que la última columna es la etiqueta y las demás son características
X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values

# Convertir las etiquetas de texto a números
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)

# Convertir las etiquetas a una codificación one-hot
y = to_categorical(y)

# Dividir los datos en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Escalar las características
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Crear el modelo de la red neuronal
model = Sequential()

# Añadir la capa de entrada con 7 neuronas (una por cada característica de entrada)
model.add(Dense(7, input_dim=X_train.shape[1], activation='tanh'))

# Añadir la capa oculta con 4 neuronas y activación 'tanh'
model.add(Dense(4, activation='tanh'))

# Añadir la capa de salida con 3 neuronas (una por cada clase) y activación 'sigmoid'
model.add(Dense(3, activation='sigmoid'))

# Compilar el modelo
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Imprimir el resumen del modelo para ver los pesos y bias
model.summary()

# Entrenar el modelo
model.fit(X_train, y_train, epochs=100, batch_size=10, verbose=0, validation_data=(X_test, y_test))

# Evaluar el modelo
loss, accuracy = model.evaluate(X_test, y_test)
print(f'Pérdida: {loss}, Precisión: {accuracy}')

# Predecir y mostrar resultados de prueba
y_pred = model.predict(X_test)
# print('Predicciones:\n', y_pred)
```
